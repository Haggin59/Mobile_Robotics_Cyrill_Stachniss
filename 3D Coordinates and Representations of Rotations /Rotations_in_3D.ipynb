#Rotation matrices 3D

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# Rotation about Y-axis

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.set_xlim([0,5])
ax.set_ylim([0,5])
ax.set_zlim([0,5])

v1 = np.array([3,0,0])    #base vector

psi = -3.14/2
rot_y = np.array([[math.cos(psi), 0, math.sin(psi)], [0, 1, 0], [-math.sin(psi), 0, math.cos(psi)]]) #Rotation matrix about Y

v1_rot_y = np.dot(rot_y,v1)  #v1 rotated by -90 about Y axis 

ax.quiver(0,0,0, v1[0],v1[1],v1[2])
ax.quiver(0,0,0, v1_rot_y[0], v1_rot_y[1], v1_rot_y[2])
```

# Rotation aboout Z-axis

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.set_xlim([0,5])
ax.set_ylim([0,5])
ax.set_zlim([0,5])

v2 = np.array([3,0,0])    #base vector

psi = 3.14/4
rot_z = np.array([[math.cos(psi), -math.sin(psi), 0], [math.sin(psi), math.cos(psi), 0], [0, 0, 1]]) #Rotation matrix about Z

v2_rot_z = np.dot(rot_z,v2)  #v1 rotated by -90 about Z axis 

ax.quiver(0,0,0, v2[0],v2[1],v2[2])
ax.quiver(0,0,0, v2_rot_z[0], v2_rot_z[1], v2_rot_z[2])
```
