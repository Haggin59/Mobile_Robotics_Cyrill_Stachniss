##Operations_on_vectors_2D

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
```

```python

v1 = np.array([ 2, 2 ])


plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')


plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6 ))
plt.grid()
plt.show()
```

##Translation

```python
# X' = [x, y] + [tx, ty]

v1_trans = v1 + [1,2]

plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')
plt.plot([0,v1_trans[0]],[0,v1_trans[1]],'r',label = 'Translation')

plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6 ))
plt.grid()
plt.show()
```

##Rotation

```python

# X'= R(X)= [x', y'] = [[cos Ψ, -sin Ψ], [sin Ψ, cos Ψ]].[x, y]
# Ψ = angle of rotation

psi = 3.14/4
rotation_matrix = np.array([[math.cos(psi), - math.sin(psi)], [math.sin(psi), math.cos(psi)]])
v1_rot = np.dot(rotation_matrix,v1)
# print(rotation_matrix)
# print(v1_rot)

plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')
plt.plot([0,v1_rot[0]],[0,v1_rot[1]],'y',label = 'Rotation')


plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6 ))
plt.grid()
plt.show()
```

##Scale change

```python
#X' = D(X) = [x', y'] = [[ƛ, 0],[0, ƛ]].[x, y] = [ƛx, ƛy]
#ƛ = scaling factor

lam = 3
scaling_matrix = np.array([[lam, 0],[0, lam]])
v1_scaled = np.dot(scaling_matrix,v1)

plt.plot([0, v1_scaled[0]], [0, v1_scaled[1]], 'r', label = "Scaled")
plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')

#print(scaling_matrix)
#print(v1_scaled)

plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6 ))
plt.grid()
plt.show()
```
